```python
import pandas as pd
```

```python
# Load Intercom CSV export
intercom_df = pd.read_csv('../csv/intercom/20180711-20200710.csv')
intercom_df.shape
```

```python
intercom_df.dtypes
```

```python
# Convert date columns to proper datetime objects
intercom_df['Created at'] =  pd.to_datetime(intercom_df['Created at'])
intercom_df['Last updated at'] =  pd.to_datetime(intercom_df['Last updated at'])
```

```python
intercom_df.head(3)
```

```python
intercom_df.tail(3)
```

```python
# Set the "Created at" date as the data frame's index
intercom_df = intercom_df.set_index('Created at')
intercom_df.head(3)
```

```python
intercom_df.index
```

```python
# Add columns with year, month, weekday name, and quarter
intercom_df['Year'] = intercom_df.index.year
intercom_df['Quarter'] = intercom_df.index.quarter
intercom_df['Month'] = intercom_df.index.month
intercom_df['Weekday Name'] = intercom_df.index.day_name()
intercom_df['year_month'] = intercom_df.index.strftime('%Y-%m')

# Display a random sampling of 5 rows
intercom_df.sample(5, random_state=0)
```

```python
# Time-based indexing allows for handy stuff like this:

# Select a single day
day = intercom_df.loc['2018-08-10']
```

```python
# Get the number of records for this day
number_of_rows = len(day)
print(number_of_rows)
```

```python
# Select a slice of days
intercom_df.loc['2020-06-30':'2020-07-10']
```

```python
# Partial-string indexing lets us select on a partial match, neat
intercom_df.loc['2019-01']
```
